```python
def countdown():
  i = 5
  while i > 0:
    yield i
    i -= 1

for i in countdown():
  print(i)
```

```python
def numbers(x):
  for i in range(x):
    if i % 2 == 0:
      yield i

print(list(numbers(11)))
```

```python
def make_word():
  word = ""
  for ch in "spam":
    word +=ch
    yield word

print(list(make_word()))
```

```python
def decor(func):
  def wrap():
    print("==========")
    func()
    print("==========")
  return wrap

def print_text():
  print("Hello world!")

decorated = decor(print_text)
decorated()
```

```python
def factorial(x):
  if x == 1:
    return 1
  else:
    return x * factorial(x - 1)

print(factorial(5))
```

```python
def fib(x):
  if x == 0 or x == 1:
    return 1
  else:
    return fib(x-1) + fib(x-2)
print(fib(4))
```

```python
def power(x, y):
  if y == 0:
    return 1
  else:
    return x * power(x, y-1)

print(power(2, 3))
```
